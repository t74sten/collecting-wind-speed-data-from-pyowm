```python
import pyowm
import pandas as pd
import time
import datetime

from pyowm import OWM
from pyowm.utils import timestamps, formatting
```

Код сверху осуществляет импорт необходимых библиотек для написания нашей программы.
Pyowm является библиотекой портала open weather map позволяющей настроить среду на работу программы с порталом.
Pandas - библиотека для работы с массивами данных.
Библиотеки time и datetime необходимы для формирования списка с датой временем.

```python
wind_param = []
date_time = []
```

Далее объявляем списки для хранения запрашиваемой информации информации. 
wind_param - для хранения параметров ветра (скорость и угол).
date_time - для хранения даты в ремени в формате: день, месяц, год, часы, минуты, секунды.

```python
while len(wind_param) < 10000:
    date = datetime.datetime.now().strftime("%d-%m-%Y %H:%M:%S")
    date_time.append(date)
    owm = OWM('c7a85f9516114a594e0cf66942c2aa65')
    mgr = owm.weather_manager()
    observation = mgr.weather_at_place('Voronezh,RU')
    w = observation.weather
    wind_param.append(w.wind())
    print(w.wind(), date)
    time.sleep(300)
```

Выше приведен основной код осуществляющий настройку передачи информации между программой и серверами pyowm.
Представлен цикл while, который будет повторяться, пока не наступит определенное событие, в нашем случае событием будет являться величина списка. Так как величина списка и будет количеством строк в нашем дата сете мы можем изменять этот параметр по нашим пожеланиям и создавать датасеты необходимой величины.

Внутри цикла while создается переменная date для запроса текущей даты и времени и следующей строкой эти данные передаются в ранее созданный список date_time.

Далее внутри цикла осуществляется импорт объекта OWM из библиотеки для ввода API ключа, который разрешает доступ к серверам OWM. 
Данный ключ можно получить бесплатно при регистрации на сайте openweathermap. 
Далле идет настройка параметров сообщения с серверами, вводится местоположение для которого требуется узнать параметры и обозначается переменная, ответственная за запрос (w). Следующей строкой кода эта информация передается в список wind_param.

Параметр в скобках для функции time.sleep(1) - это время (в секундах) выдержки цикла перед перезапуском, изменяя данный параметр мы можем формировать временной интервал для сбора датасета, например, делать запрос каждые 5 или 300 секунд."

```python
wind_param_DF = pd.DataFrame(wind_param)
date_time_DF = pd.DataFrame({'date_time': date_time})
DF = pd.concat([date_time_DF,wind_param_DF],1).reset_index(level=0,drop=True)
DF.to_excel("wind_data_voronezh_11.05.2021-18.05.2021.xlsx")
DF
```

После того, как выполнение программы закончится, необходимо сформировать общий набор данных из скорости, направления ветра и времени. В этом на поможет библиотека Pandas. С помощью её функций мы приводим список с датами и временем из формата списка в формат словаря и объединяем со словарем, содержащим параметры со скоростью ветра и направлением. Далее производим сохранение в файл с расширение xlsx. Pandas позволяет сохранить данные в любом формате с любым расширением, поддерживаемым ОС Windows. После преобразований дата сет сформирован и мы можем проводить дальнейшие исследования данных и работу с ними в любой удобной программе.

# РАБОТА С ДАННЫМИ

```python
DF['date_time'] = pd.to_datetime(DF['date_time'])
DF = (DF.groupby(['deg',pd.Grouper(freq='W-MON', key='date_time')])
        .size()
        .unstack(fill_value=0)
        .sort_index(axis=1))

DF.columns = DF.columns.date
DF['Total_frequency'] = DF.sum(axis=1)
print(DF)
```
